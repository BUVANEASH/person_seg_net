# Converting from tf to Core-ML
#### Activate a python environment with tf==1.13.1 cpu before running this notebook, tfcoreml requires tf-cpu.

```python
import tfcoreml as tf_converter
import os
MODEL_PATH = '/media/new_hdd2/seg_net/results/Models/Models'
# Output protocol buffer model path
pb_filename = 'seg_net.pb'
mlmodel_filename = 'seg_net.mlmodel'
```

```python
tf_converter.convert(tf_model_path = os.path.join(MODEL_PATH,pb_filename),
                     mlmodel_path = os.path.join(MODEL_PATH,mlmodel_filename),
                     output_feature_names = ['import/sigmoid/Sigmoid:0'],
                    image_input_names=['import/data:0'],
                    input_name_shape_dict={'import/data:0':[1, 768, 768, 3]},
                    image_scale=1/255.0)
```
